```python
import pandas as pd
import os
from tqdm import tqdm
import uuid
import math
import requests
```

```python
contact_file = pd.read_csv('/Users/stephanieolaiya/Downloads/paper-contacts.final.csv',index_col=0)
contact_file
```

```python
# get all papers that have gene sets in rummgene
def getRummagenePapers():
  RUMMAGENE_URL = 'https://rummagene.com/graphql'
  json= """{
    pmcInfos {
      nodes {
        pmcid
      }
    }
  }"""

  response = requests.post(RUMMAGENE_URL, json={"query": json}).json()
  rummagene_papers = [node_item['pmcid'] for node_item in response['data']['pmcInfos']['nodes']]
  return rummagene_papers
```

```python
rummagene_papers = getRummagenePapers()
```

```python
len(rummagene_papers)
```

```python
filtered_papers = contact_file[contact_file.index.isin(rummagene_papers)]
filtered_papers
```
